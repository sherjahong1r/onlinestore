# **2. Lug‘at yaratish va matnni kodlash**





## Barcha sharhlardagi so‘zlarni tokenize qiling.



## collections.Counter yordamida so‘zlar chastotasini hisoblang.



## Eng ko‘p uchraydigan 3000 ta so‘zdan iborat lug‘at (vocab) yarating.



## Lug‘atga <pad> (indeks 0) va <unk> (indeks 1) maxsus tokenlarini qo‘shing.



## Matnni sonli ketma-ketlikka o‘giruvchi encode_text funksiyasini yarating.

```python
import torch
from collections import Counter
# torch chuqur o'rganish uchun, Counter esa elementlarning chastotasini hisoblash uchun ishlatiladi.

def tokenize(text):
  return text.lower().split()
# Berilgan matnni kichik harflarga o'tkazadi (lower()) va so'zlarga ajratadi (split()).
# Misol uchun, "Hello World" -> ['hello', 'world'].

counter = Counter(word for text in df['text'] for word in tokenize(text))
# df['text'] ustunidagi barcha sharhlardagi so'zlarning necha marta uchraganini
# hisoblab chiqadi. Bu kod barcha so'zlarni yig'ib, ularning sonini sanaydi.
most_common = counter.most_common(3000) # Eng ko'p uchraydigan 3000 so'zni va ularning sonini oladi
vocab = {word: i+2 for i, (word, _) in enumerate(most_common)}
vocab['<pad>'] = 0
vocab['<unk>'] = 1
# Tanlangan 3000 ta eng ko'p uchraydigan so'zlar uchun lug'at (vocabulary) yaratadi.
# Har bir so'zga noyob raqamli indeks beradi. <pad> (to'ldirish) uchun 0 va <unk> (noma'lum so'zlar)
# uchun 1 indekslari qo'shiladi. Bu NLP modellarida matnni raqamli ko'rinishga o'tkazishda muhim.

def encode(text):
  return torch.tensor([vocab.get(word, 1) for word in tokenize(text)])
# encode(text) funksiyasi: Berilgan matndagi so'zlarni yuqorida yaratilgan vocab lug'ati
# yordamida raqamli indekslarga o'tkazadi. Agar so'z lug'atda bo'lmasa, unga <unk> (1)
# indeksi beriladi. Bu matnni modelga kiritishdan oldin "raqamlashtirish" jarayonidir.
```

```python
from torch.utils.data import Dataset, DataLoader
# PyTorch kutubxonasidan Dataset va DataLoader sinflarini import qiladi. Dataset o'zingizning
# ma'lumotlaringizni PyTorch formatiga moslash uchun asosiy sinf bo'lsa, DataLoader ma'lumotlarni
# modelga kichik "paketlar" (batch) ko'rinishida yetkazib berish uchun ishlatiladi.

class IMDBDataset(Dataset):
  def __init__(self, df):
    self.X = [encode(text) for text in df['text']]
    self.y = [torch.tensor(1.0 if s == 'positive' else 0.0) for s in df['stars']]

  def __len__(self):
    return len(self.X)

  def __getitem__(self, idx):
    return self.X[idx], self.y[idx]

# class IMDBDataset(Dataset):: Bu IMDBDataset nomli yangi sinfni aniqlaydi. U Dataset sinfidan meros oladi,
# ya'ni PyTorch tomonidan tan olinadigan maxsus ma'lumotlar to'plami hisoblanadi.

# def __init__(self, df):: Bu sinfning konstruktori bo'lib, IMDBDataset obyekti yaratilganda
# avtomatik ravishda ishga tushadi. U df (DataFrame) qabul qiladi:

# self.X = [encode(text) for text in df['Reviews']]: Bu qator df dagi har bir kino sharhini (Reviews ustuni)
# encode funksiyasi yordamida raqamli indekslarga aylantiradi (ya'ni, so'zlarni raqamlarga o'giradi) va natijalarni
# self.X ro'yxatiga saqlaydi. self.X datasetdagi kirish ma'lumotlari bo'ladi.
# self.y = [torch.tensor(1.0 if s == 'positive' else 0.0) for s in df['stars']]: Bu qator df dagi har bir sharhning
# sentimentini (Sentiment ustuni, 'positive' yoki 'negative') PyTorch tensorlariga aylantiradi. 'positive' 1.0 ga, 'negative'
# 0.0 ga teng bo'ladi. self.y datasetdagi chiqish ma'lumotlari (yoki maqsad vektorlari) bo'ladi.
# def __len__(self):: Bu metod datasetdagi umumiy elementlar sonini (ya'ni, sharhlar sonini) qaytarishi shart.

# return len(self.X): Datasetdagi elementlar sonini self.X ro'yxatining uzunligi orqali hisoblaydi va qaytaradi.
# def __getitem__(self, idx):: Bu metod berilgan idx (indeks) bo'yicha datasetdan bitta elementni (sharh va uning sentimentini) qaytarishi shart.

# return self.X[idx], self.y[idx]: idx-chi sharhning raqamli ko'rinishini (self.X[idx]) va uning sentimentini
# (self.y[idx]) qaytaradi. Bu PyTorch DataLoader uchun har bir ma'lumotni olish usuli.
```

```python
def collate_fn(batch):
  Xs, ys = zip(*batch)
  # Ketma-ketliklarni uzunligi bo'yicha kamayish tartibida tartiblash
  lengths = torch.tensor([len(x) for x in Xs])
  lengths, sorted_idx = lengths.sort(descending=True)
  Xs_sorted = [Xs[i] for i in sorted_idx]
  ys_sorted = [ys[i] for i in sorted_idx]

  max_len = max(len(x) for x in Xs_sorted)
  padded_Xs = [torch.cat([x, torch.zeros(max_len - len(x), dtype=x.dtype)], dim=0) for x in Xs_sorted]

  return torch.stack(padded_Xs), torch.stack(ys_sorted), lengths

train_loader = DataLoader(IMDBDataset(df), batch_size=32, shuffle=True, collate_fn=collate_fn)
```

# **3. Modelni qurish va sozlash**





## SentimentRNN nomli Pytorch modelini yarating.



## Model arxitekturasi quyidagicha bo‘lsin:

## – embedding_dim: 100

## – hidden_dim: 256

## 3. Modelni, nn.BCELoss yo‘qotish funksiyasini va Adam optimizatorini lr=0.002 bilan sozlang.

```python
import torch.nn as nn
import torch.nn.utils.rnn as rnn_utils

class SentimentRNN(nn.Module):
  def __init__(self, vocab_size, embedding_dim=100, hidden_dim=256, dropout_prob=0.5):
    super().__init__()
    self.embedding = nn.Embedding(vocab_size, embedding_dim)
    self.dropout = nn.Dropout(dropout_prob)
    self.lstm = nn.LSTM(embedding_dim, hidden_dim, batch_first=True)
    self.fc = nn.Linear(hidden_dim, 1)
    self.sigmoid = nn.Sigmoid()

  def forward(self, x, lengths):
    x = self.embedding(x)
    x = self.dropout(x)

    # Ketma-ketliklarni packing qilish
    packed_x = rnn_utils.pack_padded_sequence(x, lengths.cpu(), batch_first=True)

    # LSTMga berish
    packed_output, (h_n, c_n) = self.lstm(packed_x)

    # Faqat oxirgi hidden state'ni ishlatamiz
    x = self.fc(h_n.squeeze(0))
    return self.sigmoid(x)
```

```python
len(vocab)
# # bu lug'atdagi (vocabulary) so'zlar sonini qaytaradi, ya'ni model ishlatadigan barcha noyob so'zlar va
# qo'shimcha maxsus tokenlar (<pad>, <unk>) sonini ko'rsatadi.
```

```python
model = SentimentRNN(len(vocab), dropout_prob=0.5)
criterion = nn.BCELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.0001)
```

# **4. Modelni o‘rgatish va bashorat qilish**





## Modelni 15 ta epoxa davomida o‘rgating va har bir epoxadagi o‘rtacha yo‘qotishni ekranga chiqaring.



## Restoran sharhlarini qabul qilib, ularning sentimentini ("positive" yoki "negative") qaytaruvchi predict_review funksiyasini yarating.




```python
# @title
for epoch in range(15):
  total_loss = 0

  for X, y, lengths in train_loader: # lengths ham o'qiladi
    y_pred = model(X, lengths) # Modelga lengths ham beriladi
    loss = criterion(y_pred, y.unsqueeze(1))

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    total_loss += loss.item()

  print(f"Epoch {epoch+1}, Loss: {total_loss:.4f}")
```

```python
def predict(text):
  model.eval()
  with torch.no_grad():
    x = encode(text)
    # predict funksiyasida ham uzunlikni hisoblash va bitta batch qilish kerak
    length = torch.tensor([len(x)])
    x = x[:100]
    if len(x) < 100:
      x = torch.cat([x, torch.zeros(100-len(x), dtype=torch.long)])
    x = x.unsqueeze(0)

    y_pred = model(x, length) # predict funksiyasida ham lengths beriladi

    prob = y_pred.item()
    label = "positive" if prob > 0.5 else "negative"
    print(f"Ishonchi: {prob:.2f}, Bashorat: {label}")
```

```python
predict("Taomlari juda mazali, lekin xizmat ko‘rsatish juda sekin.")
predict("Bu restoranga boshqa qaytib kelmayman. Ofitsiant juda qo‘pol edi.")
predict("Hizmat ham taomlari ham juda mazzali ekan")
predict("hammasi juda juda zo'r, ayniqsa taomlari juda mazzali edi")
```

```python
model.eval()

all_labels = []
all_preds = []

with torch.no_grad():
  for X, y, lengths in train_loader: # lengths ham o'qiladi
    y_pred = model(X, lengths) # Modelga lengths ham beriladi
    predicted_labels = (y_pred > 0.5).int()

    all_labels.extend(y.cpu().numpy().flatten())
    all_preds.extend(predicted_labels.cpu().numpy().flatten())

print("Model baholash uchun barcha sharhlar bo'yicha bashoratlar to'plandi.")
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(all_labels, all_preds)

print(f"Modelning aniqligi: {accuracy:.4f}")
```
