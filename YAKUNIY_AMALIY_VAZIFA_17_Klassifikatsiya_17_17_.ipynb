# **ML ga tayyorlaymiz**

## X va y ga ajratib olamiz

```python
X = df.drop('class', axis=1).values
X[:5]
```

```python
y = df['class']
y[:5]
```

## Train/test split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

# **Model**

## from sklearn.tree import DecisionTreeClassifier

### bu kutubxona yordamida model qilindi

```python
# from sklearn.tree import DecisionTreeClassifier

tree_model = DecisionTreeClassifier()
tree_model.fit(X_train, y_train)
```

# **Baholash**

```python
y_predict = tree_model.predict(X_test)
```

```python
# from sklearn.metrics import classification_report

print(classification_report(y_test, y_predict))
```

```python
# from sklearn.metrics import accuracy_score
print("DecisionTree's Accuracy: ", accuracy_score(y_test, y_predict))

# Bu umumiy aniqligini chiqaradi
```

```python
# from sklearn.metrics import confusion_matrix

sns.heatmap(confusion_matrix(y_test, y_predict), annot=True)
plt.show()
```

```python
confusion_matrix(y_test, y_predict)
# grafiksiz ko'rinishi
```

# **Hyperparameters**

```python
# from sklearn.tree import DecisionTreeClassifier
# from sklearn import tree

tree_model = DecisionTreeClassifier(min_impurity_decrease=0.01, max_depth=5)
tree_model.fit(X_train, y_train)
y_predict = tree_model.predict(X_test)
print("DecisionTree's Accuracy: ", accuracy_score(y_test, y_predict))

cols = df.drop('class', axis=1).columns.tolist()
classes = ['edible', 'poisonous']

plt.figure(figsize=(30,20))
tree.plot_tree(tree_model, feature_names=cols, class_names=classes, filled=True)
plt.show()
```

##Random Forest

```python
# from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(n_estimators=20)
rf_model.fit(X_train, y_train)
y_pred = rf_model.predict(X_test)
print(classification_report(y_test, y_pred))
```

# **K-Nearest Neighbors (KNN)**

```python
# from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train)
# StandardScaler bilan fit orqali ma'lumotlarni o'rganib qo'yayabmiz
```

```python
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

# scaler.transform(X_train): X_train (o'quv to'plami) dagi barcha ustunlarni standartlashtiradi.
# Bunda har bir ustun o'zining o'rtacha qiymati (mean) va standart og'ishiga
# (standard deviation) qarab o'zgartiriladi. Bu ma'lumotlarni bir xil shkalaga olib keladi.
```

```python
# from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=15) # k-ni qiymati

knn.fit(X_train_scaled, y_train)
```

### Dastlab knn da n_neighbors=15 qilib ko'rdik taxminiy pastdagi kod orqali eng yaxshi k qiymatni qo'yib qaytadan kodni ishga tushiramiz

```python
# from sklearn.model_selection import GridSearchCV

param_grid = {'n_neighbors': np.arange(1, 18)}
knn_gscv = GridSearchCV(knn, param_grid, cv=5)
knn_gscv.fit(X, y)
# GridSearchCV - bu k kabi model parametrlarini (hyperparameters) avtomatik ravishda sinab ko'rib,
# modelning eng yaxshi ishlaydigan qiymatini topish uchun ishlatiladigan vositadir.
```

```python
knn_gscv.cv_results_['rank_test_score']
```

```python
knn_gscv.best_params_
# demak eng yaxshi k qiymat 1 ekan
```

```python
knn_gscv.best_score_
# bu GridSearchCV tomonidan topilgan eng yaxshi parametrlar kombinatsiyasining (bu yerda n_neighbors)
# kross-validatsiya bo'yicha erishilgan eng yuqori baho (skor) qiymatini qaytaradi. Bu modelning eng yaxshi ishlash ko'rsatkichidir.
```

```python
# import matplotlib.pyplot as plt
# import seaborn as sns
```

## Bunda eng yaxshi k qiymati eng balandga ko'tarilgani bo'ladi yani bu yerda 1 ekan va buni tushunarli bo'lishi uchun grafik orqali ta'svirladik

```python
plt.figure(figsize=(10,6))
plt.plot(param_grid['n_neighbors'], knn_gscv.cv_results_['mean_test_score'])
# n_neighbors (ya'ni, k qiymatlari) bo'yicha mean_test_score (o'rtacha test skori) ni aks ettiruvchi chiziqli grafik chizadi.
plt.xticks(param_grid['n_neighbors'], rotation=90)
# Bu qator x o'qi bo'ylab belgilar (ticks) sifatida n_neighbors qiymatlarini ko'rsatishni ta'minlaydi, bu esa grafikni o'qishni osonlashtiradi.
plt.xlabel("k")
plt.ylabel("Xatolik reytingi")
plt.grid()
plt.show()
```

```python
# from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=1) # k-ni qiymati

knn.fit(X_train_scaled, y_train)

# k ning eng yaxshi varianti 1 ekanligini yuqoridan aniqladik va qaytadan run qilyabmiz
```

```python
y_predict = knn.predict(X_test_scaled)
# Ushbu qator tayyorlangan k-NN modeli yordamida sinov ma'lumotlaridagi (test data) y qiymatlarini bashorat qilish uchun ishlatiladi.
```

## Yuqorida k qiymatini aniqladik va qo'yib pastda natijasini yana tekshirib ko'ramiz

## Confusion matrix yordamida tekshiramiz

```python
# from sklearn.metrics import confusion_matrix
# import seaborn as sns
# import matplotlib.pyplot as plt

sns.heatmap(confusion_matrix(y_test, y_predict), annot=True)
plt.show()
```

```python
# from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, y_predict)

print("RMSE=", np.sqrt(mse))
```

## **Logistic Regression**

```python
# from sklearn.linear_model import LogisticRegression

# Logistic Regression modelini yaratish
log_reg_model = LogisticRegression(solver='liblinear', random_state=42)

# Modelni o'qitish
log_reg_model.fit(X_train, y_train)

# Bashorat qilish
y_pred_log_reg = log_reg_model.predict(X_test)

# Modelni baholash
print(classification_report(y_test, y_pred_log_reg))

print("Logistic Regression Accuracy: ", accuracy_score(y_test, y_pred_log_reg))
```

Logistic Regression

classification_report bu klassifikatsiya modelingizning har bir sinf bo'yicha qanchalik samarali ishlaganini ko'rsatadigan baholash hisobotidir.

U quyidagi asosiy ko'rsatkichlarni o'z ichiga oladi:

Precision (Aniqlik): Bashorat qilingan ijobiy holatlarning qanchasi haqiqatan ham ijobiy.

Recall (To'liqlik): Haqiqiy ijobiy holatlarning qanchasi to'g'ri aniqlangan.

F1-Score (F1-ko'rsatkich): Precision va Recall ning muvozanatlashgan o'rtachasi.

Accuracy (Umumiy aniqlik): Barcha to'g'ri bashoratlarning umumiy foizi.

Bu hisobot modelingizning har bir sinfda qanchalik yaxshi ishlaganini tushunishga yordam beradi.

Logistic Regression modelining **Confusion Matrix**'ini ko'ramiz.

```python
# import seaborn as sns
# import matplotlib.pyplot as plt
# from sklearn.metrics import confusion_matrix

plt.figure(figsize=(6, 5))
sns.heatmap(confusion_matrix(y_test, y_pred_log_reg), annot=True, fmt='d', cmap='Blues')
plt.title('Logistic Regression Confusion Matrix')
plt.xlabel('Bashorat qilingan qiymat')
plt.ylabel('Haqiqiy qiymat')
plt.show()
```
