## *Klasslar nomutanosibligini bartaraf etish SMOTE metodidan foydalanamiz*

## *`jiddiy_kechikish_2yil` ustunida klasslar nomutanosibligi mavjud (0 qiymatlari 1 qiymatlariga nisbatan ko'proq). SMOTE yordamida ozchilik klassini ko'paytiramiz. Yani ikkalsining qiymatlari tenglashtiriladi bu qurmoqchi bo'lgan modelimiz aniqligini oshiradi*

```python
from imblearn.over_sampling import SMOTE
from collections import Counter

# imblearn.over_sampling.SMOTE: Bu algoritmni qo'llash uchun ishlatiladi. Asosiy vazifasi: ma'lumotlar to'plamidagi sinflar o'rtasidagi
# nomutanosiblikni (imbalansni) bartaraf etish. Ya'ni, kichikroq sinfdagi ma'lumotlarni sun'iy ravishda ko'paytirib, sinflarni muvozanatlashtiradi.
# collections.Counter: Bu elementlarning takrorlanishini hisoblash uchun ishlatiladi. Qaysi qiymatdan nechta borligini sanab, lug'at
# (dictionary) ko'rinishida qaytaradi. Masalan, bizning holatimizda jiddiy_kechikish_2yil ustunidagi 0 va 1 qiymatlari necha marta uchraganini hisoblaydi.

X = df.drop('jiddiy_kechikish_2yil', axis=1)
y = df['jiddiy_kechikish_2yil']

print(f'Asl maqsadli taqsimot yani ustun: {Counter(y)}')

sm = SMOTE(random_state=42)
X_res, y_res = sm.fit_resample(X, y)

print(f'Qayta namunalangan maqsadli taqsimot: {Counter(y_res)}')

df_resampled = pd.concat([X_res, y_res], axis=1)

# U ozchilik klassiga (bizning holatimizda jiddiy_kechikish_2yil = 1 bo'lgan qatorlarga) tegishli namunalarni topadi.
# Har bir ozchilik klassi namunasi uchun uning eng yaqin qo'shnilarini (o'xshash namunalarini) aniqlaydi.
# Asl namuna va uning qo'shnilari orasidagi farqlarni interpolation usuli bilan hisoblab, yangi, sun'iy namunalar hosil qiladi.
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import  confusion_matrix, classification_report, accuracy_score
import sklearn.metrics as metrics

# Modelni yaratamiz (training)
RF_model = RandomForestClassifier(n_estimators=9)
RF_model.fit(X_train, y_train)

# Modelni baholaymiz
y_pred = RF_model.predict(X_test)
print(classification_report(y_test, y_pred))
print("Model aniqligi:", accuracy_score(y_test,y_pred))

## confusion matrix
conf_mat = confusion_matrix(y_test, y_pred)
sns.heatmap(conf_mat, annot=True,fmt="g")
plt.show()

## ROC curve
fpr, tpr, thresholds = metrics.roc_curve(y_test, y_pred)
roc_auc = metrics.auc(fpr, tpr)
display = metrics.RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc, estimator_name='ROC curve')
display.plot()
plt.show()
```
